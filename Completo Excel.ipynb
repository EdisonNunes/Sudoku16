```python
from openpyxl import load_workbook
import openpyxl
from openpyxl.worksheet.copier import WorksheetCopy
from openpyxl.styles import Alignment, Font
import pandas as pd
from pathlib import Path
```

```python
# Fonte ft1  Tamanho 22 Negrito Preto
# Fonte ft2  Tamanho 11 Normal  Azul
# Fonte ft3  Tamanho 22 Negrito Azul

ft1 = Font(size=22, bold=True )  # Preto Negrito
ft2 = Font(size=11, bold=False, color='0000FF')  # Azul Normal
ft3 = Font(size=22, bold=True, color='0000FF')   # Azul Negrito
alignment=Alignment(horizontal='center', vertical='center')
caminho = Path.cwd() / Path('Planilhas')
```

```python
# Carrega a planilha do jogo

workbook = openpyxl.load_workbook(caminho / Path('Sudoku16.xlsx'))

ws = workbook.active
nomes_guias = workbook.sheetnames
ultima_guia = nomes_guias[len(nomes_guias)-1]
if ultima_guia == 'Modelo':
    ultima_guia = 0
guia_atual = int(ultima_guia)+1
sheet_name = str(guia_atual)
if len(sheet_name) < 2:
    sheet_name = '0' + sheet_name
    
# cria uma nova guia

template_worksheet = workbook['Modelo']
new_worksheet      = workbook.create_sheet(sheet_name)
instance           = WorksheetCopy(template_worksheet, new_worksheet)
WorksheetCopy.copy_worksheet(instance)


for i, s in enumerate(nomes_guias):
    if s == sheet_name:
         break

ws = workbook.active
workbook.active = i+1

#print(nomes_guias)
#print(ultima_guia)
#print(guia_atual)
#print(sheet_name)
#print(i+1)

```

```python
# Le planilha Matriz com o jogo
# Transforma planilha lida em uma lista de listas (frame) com os valores das colunas lidas na Matriz

#df = pd.read_excel('Matrix_16.xlsx')
df = pd.read_excel(caminho / Path('Matrix_16.xlsx'))

frame = []
for column in df.columns:
    # Storing the rows of a column into a temporary list
    li = df[column].tolist()
    
    # appending the temporary list
    frame.append(li)
    
#print(frame)  
```

```python
# Cria uma lista de listas (jogo) conforme o modelo 

jogo = []
li = []
for i in range(0, 16):
    #print(frame[0][i], frame[i][0] )
    for j in range(0, 16):
        li.append(frame[j][i]) 
        
    jogo.append(li) 
    li = []
    
#print(jogo)
```

```python
# Troca '0' por espaço

#sheet = workbook.active
sheet = new_worksheet

for i, linha in enumerate(jogo):
    #print(f'linha ........ = {linha}')
    for j, cell in  enumerate(linha):
        #print(f'i= {i} cell = {cell}  j= {j}') 
        if (cell == '0') or (cell == 0):
            linha[j] = ''
            preenchido = False
        else:
            linha[j] = cell
            preenchido = True
            
        if preenchido:
            sheet.cell(column=j+1, row=i+1, value=linha[j]).font = ft1  
            
```

```python
# Função procura_quadrante
# Parâmetros : Linha da Planilha
#              Coluna da Planilha
# Resultado:
#             quadrante da planilha em que está a célula(linha, coluna)


def procura_quadrante(linha, coluna):
    if (linha < 5 ) and (coluna < 5 ):
        quadrante = 1
    elif (linha < 5 ) and (coluna >= 5 and coluna < 9):
        quadrante = 2
    elif (linha < 5 ) and (coluna >= 9 and coluna < 13):
        quadrante = 3
    elif (linha < 5 ) and (coluna >= 13):
        quadrante = 4
    elif (linha >= 5 and linha < 9) and (coluna < 5 ):
        quadrante = 5
    elif (linha >= 5 and linha < 9) and (coluna >= 5 and coluna < 9):
        quadrante = 6   
    elif (linha >= 5 and linha < 9) and (coluna >= 9 and coluna < 13):
        quadrante = 7 
    elif (linha >= 5 and linha < 9) and (coluna >= 13):
        quadrante = 8
    elif (linha >= 9 and linha < 13) and (coluna < 5 ):
        quadrante = 9 
    elif (linha >= 9 and linha < 13) and (coluna >= 5 and coluna < 9):
        quadrante = 10  
    elif (linha >= 9 and linha < 13) and (coluna >= 9 and coluna < 13):
        quadrante = 11  
    elif (linha >= 9 and linha < 13) and (coluna >= 13):
        quadrante = 12 
    elif (linha >= 13) and (coluna < 5 ):
        quadrante = 13
    elif (linha >= 13) and  (coluna >= 5 and coluna < 9):
        quadrante = 14 
    elif (linha >= 13) and  (coluna >= 9 and coluna < 13):
        quadrante = 15    
    else:
        quadrante = 16
    return quadrante    


# Função qual_quadrante
# Parâmetros : quadrante da Planilha
# Resultado:
#              número da linha (x) e número da coluna (y) do início do quadrante


def qual_quadrante(quadrante):
    if quadrante == 1:
        x = 1
        y = 1
    elif quadrante == 2:
        x = 1
        y = 5
    elif quadrante == 3:
        x = 1
        y = 9 
    elif quadrante == 4:
        x = 1
        y = 13     
    elif quadrante == 5:
        x = 5
        y = 1
    elif quadrante == 6:
        x = 5
        y = 5
    elif quadrante == 7:
        x = 5
        y = 9 
    elif quadrante == 8:
        x = 5
        y = 13
    elif quadrante == 9:
        x = 9
        y = 1
    elif quadrante == 10:
        x = 9
        y = 5
    elif quadrante == 11:
        x = 9
        y = 9 
    elif quadrante == 12:
        x = 9
        y = 13  
    elif quadrante == 13:
        x = 13
        y = 1
    elif quadrante == 14:
        x = 13
        y = 5
    elif quadrante == 15:
        x = 13
        y = 9 
    else:
        x = 13
        y = 13  
        
    return x, y    

# Função existe_no_quadrante
# Parâmetros : quadrante da Planilha
#              número a ser localizado
#              lista com o jogo completo
# Resultado:
#              True se número localizado
#              False se número não localizado
        
def existe_no_quadrante(quadrante, num, lista):
        
    # procura numero no quadrante
    
    #print(jogo)
    x, y = qual_quadrante(quadrante)
    for i in range(x, x + 4):
        for j in range(y, y + 4):
            #print(f'linha= {i}  coluna= {j} valor= {lista[i-1][j-1]}')
            
            if str(num) == str(lista[i-1][j-1]):  # se existe esse número no quadrante 
                return True
                break
    return False 

# Função existe_horizontal_vertical
# Verifica se o número (num) já está presente na linha ou coluna
# Parâmetros : linha a ser pesquisada
#              Coluna a ser pesquisada
#              número a ser localizado
#              lista com o jogo completo
# Resultado:
#              True se número já existe na linha ou coluna
#              False se número não existe na linha ou coluna

def existe_horizontal_vertical(linha, coluna, num, lista):
    
    for i in range(1,17):
        #print(f'Existe {num} em ......... {lista[linha-1][i-1]} ?')  # se existe na horizontal
        #K = input("Press Enter to continue: ")
        if str(num) == str(lista[linha-1][i-1]):  # se existe na horizontal
            return True
    for j in range(1,17):
        # print(num, lista[j-1][coluna-1])  # se existe na vertical
        if str(num) == str(lista[j-1][coluna-1]):  # se existe na vertical
            return True    
        
    return  False   
         
```

```python
# Preeche o jogo com as possibilidades de preenchimento de cada célula
# Salva Planilha

lista = []
jogo_real = jogo
for i, linha in enumerate(jogo):
    #print(f'linha ........ = {linha}')
    for j, cell in  enumerate(linha):
        #print(f'i= {i}  j= {j}  cell = {cell}') 
        if cell == '':
            valor = ''
            for var in range(1,17):
                q = procura_quadrante(i+1, j+1)
                if var == 16:
                    num = 'G'
                else:
                    num = hex(var)[-1:].upper()
                    
    
                if not existe_no_quadrante(q, num, jogo):
                    if not existe_horizontal_vertical(i+1, j+1, num, jogo):
                        valor += num
                        

            if len(valor) > 1:
                sheet.cell(column=j+1, row=i+1, value=valor).font = ft2
            else:
                texto = ''
                sheet.cell(column=j+1, row=i+1, value=valor).font = ft3
                texto = str(valor) + ',' + str(q) + ',' + str(i+1) + ',' + str(j+1)
                #print(f'i= {i}  j= {j}  cell = {valor}')
                lista.append(texto.split(','))
                

                
            jogo_real[i][j]= valor
        else:
            jogo_real[i][j]= cell

#workbook.save('Sudoku16.xlsx')
workbook.save(caminho / Path('Sudoku16.xlsx'))
#print(lista)
#print(jogo_real)
```

```python
# Função retira_linha
# Percorre uma linha do jogo para procurar se existe um valor específico 
# Se encontrado, retira da lista
# Parâmetros : item com o valor de uma célula que só possui um número no preenchimento (diferente do modelo inicial)
#              lista com a linha do jogo a ser verificada
# Resultado:
#             nova_linha com o valor retirado da lista


def retira_linha(item, lista):
    valor, q, linha, coluna = item
    novo_valor =''
    nova_linha = []
    #print(valor, q, linha, coluna)
    for i in range(1,17):
        valores = lista[int(linha)-1][i-1]
        novo_valor = valores
        if len(valores) > 1:
            #print(i, valores)
            if valor in valores:
                for x in range(len(valor)):
                    novo_valor = novo_valor.replace(valor[x],"")
        #print(valores, novo_valor)
        nova_linha.append(novo_valor)

        
    #print(f'linha atual: {jogo_real[int(linha)-1]}') 
    #print(f'linha modificada: {nova_linha}')
    return(nova_linha)


# Função retira_coluna
# Percorre uma coluna do jogo para procurar se existe um valor específico 
# Se encontrado, retira da lista
# Parâmetros : item com o valor de uma célula que só possui um número no preenchimento (diferente do modelo inicial)
#              lista com a linha do jogo a ser verificada
# Resultado:
#             nova_coluna com o valor retirado da lista

def retira_coluna(item, lista):
    valor, q, linha, coluna = item
    novo_valor =''
    nova_coluna = []
    #print(valor, q, linha, coluna)
    for j in range(1,17):
        valores = lista[j-1][int(coluna)-1]
        novo_valor = valores
        if len(valores) > 1:
            #print(j, valores)
            if valor in valores:
                for x in range(len(valor)):
                    novo_valor = novo_valor.replace(valor[x],"")
        #print(valores, novo_valor)
        nova_coluna.append(novo_valor)

        
    
    #print(f'coluna modificada: {nova_coluna}')
    return(nova_coluna)

# Função retira_quadrante
# Percorre o quadrante indicado do jogo para procurar se existe um valor específico 
# Se encontrado, retira do quadrante
# Parâmetros : quadrante a procurar
#              valor a ser encontrado
# Resultado:
#             Jogo corrigido

def retira_quadrante(q, valor):
    x, y = qual_quadrante(q)
    #print(f'x = {x}  y = {y}')
    novo_valor = ''
    for i in range(x, x + 4):
        for j in range(y, y + 4):
            celula = jogo_real[i-1][j-1]
            #print(f'linha= {i}  coluna= {j} valor= {valor} celula = {jogo_real[i-1][j-1]}')
            if valor in celula:  # se existe esse número no quadrante
                if len(celula) > 1:
                    for x in range(len(celula)-1):
                        if valor == celula[x]:
                            celula = celula.replace(celula[x],"")
                            jogo_real[i-1][j-1] = celula
                    #print(f'linha= {i}  coluna= {j} valor= {valor} celula nova = {jogo_real[i-1][j-1]}')
    
                    
                

```

```python
# Varre a lista em que existem somente 1 número na célula
# Retira os valores iguais que estão na mesma linha x coluna
# Retira os valores iguais que estão no mesmo quadrante

if len(lista) > 0:
    for item in lista:
        valor, q, linha, coluna = item
        #print(item)

        retira_quadrante(int(q), valor)

        nova_linha = retira_linha(item, jogo_real)
        jogo_real [int(linha)-1] = nova_linha

        nova_coluna = retira_coluna(item, jogo_real)
        for i in range(0, 16):
            jogo_real[i][int(coluna)-1] = nova_coluna[i]
            #k = f'i = {i}  coluna = {coluna} jogo_real[i][coluna] = {jogo_real[int(i)][int(coluna)]}'
            #print(k)

```
